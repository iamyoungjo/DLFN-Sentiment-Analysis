```python
def pretty_print_review_and_label(i):
    print(labels[i] + "\t:\t" + reviews[i][:80] + "...")

g = open('reviews.txt','r') # What we know!
reviews = list(map(lambda x:x[:-1],g.readlines()))
g.close()

g = open('labels.txt','r') # What we WANT to know!
labels = list(map(lambda x:x[:-1].upper(),g.readlines()))
g.close()
```

```python
from collections import Counter
import numpy as np

positive_counts = Counter()
negative_counts = Counter()
total_counts = Counter()

pred_label = []

for i in range(len(reviews)):
    #decision = 'POSITIVE'
    for word in reviews[i].split():
        if labels[i] == 'POSITIVE':
            positive_counts[word] += 1
            total_counts[word] += 1
        else: 
            negative_counts[word] += 1
            total_counts[word] += 1


        
#         if i is 1:
# #             print(word)
#         if word == 'terrible':
#             decision = 'NEGATIVE'
#         elif word == 'improbable':
#             decision = 'NEGATIVE'
#         elif word == 'horrible':
#             decision = 'NEGATIVE'
#         elif word == 'sucks':
#             decision = 'NEGATIVE'
    
#    pred_label.append(decision)


#print(positive_counts.most_common())
# correct_count = 0
# for i in range(len(reviews)):
#     if pred_label[i] == labels[i]:
#         correct_count += 1

# print(correct_count/len(reviews))
        
```

```python
pos_neg_ratios = Counter()

for word, cnt in total_counts.items():
    if cnt > 100:
        pos_neg_ratios[word] = positive_counts[word] / (negative_counts[word]+1)
        
for word, ratio in pos_neg_ratios.items():
    if ratio > 1:
        pos_neg_ratios[word] = np.log(ratio)
    else:
        pos_neg_ratios[word] = -np.log(1/(ratio+0.01))


#print(pos_neg_ratio.most_common())
```
